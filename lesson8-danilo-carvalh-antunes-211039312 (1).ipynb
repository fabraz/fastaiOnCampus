```python
def show_img(img, title="Image"):
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
def dropout_np(img, factor=0.5):
    random_tensor = np.random.rand(*img.shape)
    dropout_mask = (random_tensor > factor).astype(np.uint8)
    return img * dropout_mask
```

```python
show_img(img, title="Original")
```

```python
dropped_img = dropout_np(img)
show_img(dropped_img, title="Dropout Factor = 0.5")
```

```python
dropped_img = dropout_np(img, 0.1)
show_img(dropped_img, title="Dropout Factor = 0.1")
```

# **Making the Kernels**

```python
filter_2 = [-0.125, 0.0, 0.125],[-0.25, 0.0, 0.25],[-0.125, 0.0, 0.125]
filter_2 = np.asarray(filter_2)
```

```python
filter_3 = [-0.125, -0.25, -0.125],[0.0, 0.0, 0.0],[0.125, 0.25, 0.125]
filter_3 = np.asarray(filter_3)
```

```python
img_4 = sig.fftconvolve(img_bw, filter_2, mode = 'same')
img_4 = np.asarray(img_4, dtype=np.uint8) / 1.25
```

```python
img_5 = sig.fftconvolve(img_bw, filter_3, mode = 'same')
img_5 = np.asarray(img_5, dtype=np.uint8) / 1.25

sobel = np.sqrt(np.square(img_4) + np.square(img_5)) / 1.25
```

The Sobel combination is a technique used for detecting edges in images. The Sobel method is an operator that calculates image intensity gradients, which are used to find the regions with the highest rates of intensity change, which often correspond to edges.

```python
fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].imshow(img_4, cmap='gray', vmin=0, vmax=255)
axs[1].imshow(img_5, cmap='gray', vmin=0, vmax=255)
axs[2].imshow(sobel, cmap='gray', vmin=0, vmax=255)
```

```python
def conv(ni, nf, ks=3, act=True):
    res = nn.Conv2d(ni, nf, stride=2, kernel_size=ks, padding=ks//2)
    if act: res = nn.Sequential(res, nn.ReLU())
    return res
```

```python
cnn = sequential(
    conv(3 ,4),
    conv(4 ,8),
    conv(8 ,16),
    conv(16,32),
    conv(32,64),
    conv(64,2, act=False),
    Flatten(),
    nn.Dropout(0.3),
)
```

- The function conv(ni, nf, ks=3, act=True) creates a convolutional layer. ni represents the number of input channels, nf is the number of filters (or output channels), ks is the kernel size (by default it is 3), and act is a boolean that indicates whether an activation function should be applied after convolution.

- cnn = sequential(...) builds multiple convolutional layers are stacked in sequence. Each call to conv adds a layer to the model. The number of input and output channels for each layer is specified, and the network increases in depth progressively.

- nn.Dropout is a regularization technique used to prevent overfitting
